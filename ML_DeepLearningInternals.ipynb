```python
# ---- Gradient calculation in Neural Networks ----
x = np.linspace(-4, 4, 100)
```

```python
def J(w):
    return 70.0 - 15.0*w**2 + 0.5*w**3 + w**4
```

```python
def dJdw(w):
    return - 30.0*w + 1.5*w**2 + 4*w**3
```

```python
plt.subplot(211)
plt.plot(x, J(x))
plt.title("J(w)")
plt.subplot(212)
plt.plot(x, dJdw(x))
plt.axhline(0, color='black')
plt.title("dJdw(w)")
plt.xlabel("w")
plt.tight_layout();
```

```python
w0 = -1
```

```python
lr = 0.001
```

```python
step = lr * dJdw(w0)
step
```

```python
w0 - step
```

```python
iterations = 50
w = w0
ws = [w]
for i in range(iterations):
  step = lr * dJdw(w)
  w -= step
  ws.append(w)
ws = np.array(ws)
```

```python
ws
```

```python
plt.plot(x, J(x))
plt.plot(ws, J(ws), 'o')
plt.plot(w0, J(w0), 'or')
plt.legend(["J(w)", "steps", "starting point"])
plt.xlim(-4.2, 0);
```
