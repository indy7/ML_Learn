```python
def mean_squared_error(y_true, y_pred):
  s = (y_true - y_pred)**2
  return s.mean()
```

```python
X = df[['Height']].values
X
```

```python
X.shape
```

```python
y_true = df['Weight'].values
y_true
```

```python
df[['Weight']].values
```

```python
y_pred = line(X)
y_pred
```

```python
y_pred.ravel()
```

```python
mean_squared_error(y_true, y_pred.ravel())
```

```python
y_pred = line(X, w=2)
mean_squared_error(y_true, y_pred.ravel())
```

```python
y_pred = line(X, w=2, b=20)
mean_squared_error(y_true, y_pred.ravel())
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD
```

```python
model = Sequential()
```

```python
model.add(Dense(1, input_shape=(1,)))
```

```python
model.summary()
```

```python
model.compile(Adam(lr=0.8), 'mean_squared_error')
```

```python
%%time
model.fit(X, y_true, epochs=40, verbose=0);
```

```python
%%time
y_pred = model.predict(X)
```

```python
df.plot(kind='scatter',
  x='Height',
  y='Weight',
  title='Weight and Height in adults')
plt.plot(X, y_pred, color='red');
```

```python
# Calculate coefficient of determination (R square)
from sklearn.metrics import r2_score
```

```python
# Train / Test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_true, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model.fit(X_train, y_train, epochs=50, verbose=0);
```

```python
y_train_pred = model.predict(X_train).ravel()
y_test_pred = model.predict(X_test).ravel()
```

```python
from sklearn.metrics import mean_squared_error as mse
```

```python
err = mse(y_train, y_train_pred)
print("Mean Squared Error (Train set):\t", "{:0.1f}".format(err))
```

```python
err = mse(y_test, y_test_pred)
print("Mean Squared Error (Test set):\t", "{:0.1f}".format(err))
```

```python
r2 = r2_score(y_train, y_train_pred)
print("R2 score (Train set):\t{:0.3f}".format(r2))
```

```python
r2 = r2_score(y_test, y_test_pred)
print("R2 score (Test set):\t{:0.3f}".format(r2))
```
